```python
import pandas as pd
import tweepy
import emoji
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader = SentimentIntensityAnalyzer()
```

```python
keys = pd.read_csv("../Clases/mis_llaves.csv", header=None)
keys = dict(zip(keys[0],keys[1]))
auth = tweepy.OAuthHandler(keys['api_key'], keys['api_secret_key'])
auth.set_access_token(keys['access_token'], keys['access_token_secret'])
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
tweets = tweepy.Cursor(api.search,
                       q=['#C√∫cuta -RT'],
                       lang='es',
                       tweet_mode='extended',
                       include_rts=False,
                       ).items(100)
```

```python
t = [tweet for tweet in tweets]
```

```python
t_processed = []

for tweet in t: 
    temp = []
    temp.append(tweet.full_text)
    temp.append(tweet.created_at)
    temp.append('C√∫cuta')
    temp.append(tweet.user.location)
    temp.append(tweet.id)
    temp.append(f"https://twitter.com/i/web/stuatus/{tweet.id}")
    t_processed.append(temp)
    
data = pd.DataFrame(t_processed)
data.columns = ['tweet', 'date', 'city', 'user_location', 'id', 'link']
data
```

```python
data = data.loc[:, ['tweet', 'date', 'link']]
data.head()
```

```python
pd.set_option('display.max_colwidth', None)
data['sent_vader'] = data['tweet'].apply(lambda val: vader.polarity_scores(val)['compound'])
```
